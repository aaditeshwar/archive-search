```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://groups.google.com/g/core-stack-nrm")

# Wait for threads list to load
#WebDriverWait(driver, 10).until(
#    EC.presence_of_element_located((By.CSS_SELECTOR, "div[role='main']"))
#)
time.sleep(5)

# Find a next link or load more button and click it
next_btn = driver.find_element(By.XPATH, "//div[@aria-label='Next page']")
next_btn.click()

time.sleep(5)

# Now retrieve the updated page source
html_after = driver.page_source
print(len(html_after))

driver.quit()
```
